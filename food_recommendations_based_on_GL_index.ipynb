```python
import pandas as pd
import numpy as np

df = pd.read_excel('/content/drive/MyDrive/ml_project/통합 식품영양성분DB_음식_20231114.xlsx')
df.head()
```

```python
subset_df = df[df['상용제품']=='품목대표']
```

```python
subset_df
```

```python
selected_columns = ['식품명', '식품대분류', '식품상세분류', '1회제공량', '에너지(㎉)', '단백질(g)', '지방(g)', '탄수화물(g)']
subset_df = subset_df[selected_columns]

subset_df.dropna(inplace=True)
```

```python
subset_df.info()
```

```python
subset_df['에너지(㎉)'] = pd.to_numeric(subset_df['에너지(㎉)'], errors='coerce')
subset_df['단백질(g)'] = pd.to_numeric(subset_df['단백질(g)'], errors='coerce')
subset_df['지방(g)'] = pd.to_numeric(subset_df['지방(g)'], errors='coerce')
subset_df['탄수화물(g)'] = pd.to_numeric(subset_df['탄수화물(g)'], errors='coerce')
```

```python
# GI지수와 GL지수 계산 함수
def calculate_gi(glucose_value, carb_value):
    return round((carb_value / glucose_value) * 100, 2)

def calculate_gl(gi_value, carb_value):
    return round((gi_value * carb_value) / 100, 2)

subset_df['GI지수'] = np.vectorize(calculate_gi)(100, subset_df['탄수화물(g)'])
subset_df['GL지수'] = np.vectorize(calculate_gl)(subset_df['GI지수'], subset_df['탄수화물(g)'])

# 결과 확인
print(subset_df.head())
```

```python
# 사용자 선호도 기반으로 전체 품목에서 추천

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import random

cat = input("식품 분류: ")

unique_subcategories = subset_df[subset_df['식품대분류'] == cat]['식품상세분류'].unique()
print("해당 분류에 속하는 식품상세분류:")
for subcategory in unique_subcategories:
    print(subcategory)

subcategory_choice = input("식품상세분류 선택: ")

food_options = subset_df[subset_df['식품상세분류'] == subcategory_choice]
random_food_choices = random.sample(list(food_options['식품명']), min(len(food_options), 5))
print("선택할 수 있는 식품:")
for food in random_food_choices:
    print(food)

food_choice = input("선택할 식품: ")

features = np.array([[food["GL지수"], food["지방(g)"], food["탄수화물(g)"], food["단백질(g)"]] for _, food in subset_df.iterrows()])
selected_food_features = subset_df[subset_df['식품명'] == food_choice][["GL지수", "지방(g)", "탄수화물(g)", "단백질(g)"]].iloc[0].to_numpy()
similarity_scores = cosine_similarity([selected_food_features], features)
top_recommendation_indices = np.argsort(similarity_scores.flatten())[::-1][1:6]

print("식품 추천 결과:")
for idx in top_recommendation_indices:
    print(subset_df.iloc[idx]["식품명"])
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import random

cat = input("식품 분류: ")

unique_subcategories = subset_df[subset_df['식품대분류'] == cat]['식품상세분류'].unique()
print("해당 분류에 속하는 식품상세분류:")
for subcategory in unique_subcategories:
    print(subcategory)

subcategory_choice = input("식품상세분류 선택: ")

food_options = subset_df[subset_df['식품상세분류'] == subcategory_choice]
random_food_choices = random.sample(list(food_options['식품명']), min(len(food_options), 5))
print("선택할 수 있는 식품:")
for food in random_food_choices:
    print(food)

food_choice = input("선택할 식품: ")

selected_food_details = subset_df[subset_df['식품명'] == food_choice]
selected_food_subcategory = selected_food_details['식품상세분류'].values[0]
selected_food_features = selected_food_details[["GL지수", "지방(g)", "탄수화물(g)", "단백질(g)"]].values
features = subset_df[["GL지수", "지방(g)", "탄수화물(g)", "단백질(g)"]].values
similarity_scores = cosine_similarity(selected_food_features, features)
gl_mask = (subset_df["GL지수"] <= 10) & (subset_df["식품상세분류"] == selected_food_subcategory)
filtered_indices = np.argsort(similarity_scores.flatten())[::-1]
filtered_indices = [idx for idx in filtered_indices if gl_mask.iloc[idx]]
top_recommendation_indices = [idx for idx in filtered_indices if subset_df.iloc[idx]["식품명"] != food_choice][:5]

print("식품 추천 결과:")
for idx in top_recommendation_indices:
    print(subset_df.iloc[idx]["식품명"])
```
